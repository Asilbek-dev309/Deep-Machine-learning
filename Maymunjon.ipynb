```
X=train.drop(['yield',],axis=1)
```

Pipeline tuzib olamiz

```
X_pipeline=Pipeline([
    ('standard_scaler',StandardScaler())
])
X_prepared=X_pipeline.fit_transform(X)
```

RandomForestRegressorni chaqirib ML qilamiz

```
RF_model=RandomForestRegressor()
RF_model.fit(X_prepared,y)
```

Predict

```
predict=RF_model.predict(X_prepared)
predict
```

Mean absolute error

```
mae=mean_absolute_error(y,predict)
mae
```

Mean squared absolute error

```
from sklearn.metrics import mean_squared_error
lin_mse = mean_squared_error(y,predict)
lin_rmse = np.sqrt(lin_mse)
print(lin_rmse)
```

Cross validation

```
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Std.dev:", scores.std())
```

```
scores = cross_val_score(RF_model, X, y, scoring="neg_mean_squared_error", cv=10)
LR_rmse_scores = np.sqrt(-scores)
display_scores(LR_rmse_scores)
```
